```python
# for 반복문 안의 코드만 수정합니다.

# --- 💡 중요: 오류 처리를 위해 NoSuchElementException을 import 해야 합니다. ---
from selenium.common.exceptions import NoSuchElementException

df = []

for url in page_url_list: # '[435]'는 예시일 뿐이므로 제거했습니다.
    driver.get(url)
    time.sleep(2)

    try:
        driver.switch_to.frame('cafe_main')

        title = driver.find_element(By.CSS_SELECTOR, '.title_text').text
        date = driver.find_element(By.CSS_SELECTOR, '.date').text[:-7] # 이 부분은 원래 코드와 동일
        content = driver.find_element(By.CSS_SELECTOR, 'div.article_container').text
        
        # --- [수정된 부분] ---
        try:
            # 댓글(review)이 있는지 먼저 시도합니다.
            # find_elements (복수) 대신 find_element (단수)를 사용해서 없으면 바로 오류를 발생시킵니다.
            review = driver.find_element(By.CSS_SELECTOR, 'div.comment_text_box > p > span').text
        except NoSuchElementException:
            # 요소를 찾지 못했다는 오류(NoSuchElementException)가 발생하면, 댓글이 없는 것이므로
            # review 변수에 '댓글 없음'이라는 값을 대신 저장합니다.
            review = '댓글 없음'
        
        # iframe에서 원래 페이지로 빠져나옵니다.
        driver.switch_to.default_content()

    except Exception as e:
        print(f"{url} 페이지 처리 중 오류 발생: {e}")
        # 오류가 나더라도 기본 프레임으로 복귀하여 다음 루프에 영향을 주지 않도록 합니다.
        driver.switch_to.default_content()
        continue # 다음 URL로 넘어갑니다.

    df.append([title, content, date, review])
```
