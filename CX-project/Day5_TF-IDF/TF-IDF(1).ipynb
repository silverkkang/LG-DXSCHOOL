```python
import pandas as pd
from kiwipiepy import Kiwi
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
df = df[df['tfidf_word'].str.len() > 30]
```

```python
df = df.reset_index(drop=True)
```

```python
text_column = 'tfidf_word'
```

```python
print("필터링 후 데이터 확인:")
print(df.head())
```

```python
# 2. 명사 추출
kiwi = Kiwi()
documents = []
for text in df[text_column].dropna():
    nouns = [token.form for token in kiwi.tokenize(text) if token.tag in {'NNG', 'NNP'}]
    documents.append(' '.join(nouns))
```

```python
# 3. 단어 수 세기 (오류가 났던 부분)
# min_df=1 로 설정해야 예제 데이터에서 오류가 나지 않습니다.
vectorizer = CountVectorizer(max_df=0.95, min_df=1) 
doc_term_matrix = vectorizer.fit_transform(documents)
```

```python
# 4. LDA 모델 학습
n_topics = 3
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(doc_term_matrix)
```

```python
def display_topics(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic #{topic_idx+1}:", " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))
```

```python
feature_names = vectorizer.get_feature_names_out()
display_topics(lda, feature_names, 5)
```
