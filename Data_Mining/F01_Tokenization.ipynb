##### ♦️1. 실습예제
  - 명사만 추출하는 define 함수를 만들어보자!

```python
def nltk_noun(string):
    split_text = nltk.word_tokenize(string)
    tagged_text = nltk.pos_tag(split_text)

    # result = []
    # for word, tag in tagged_text:
    #     if tag in {'NN', 'NNS', 'NNP', 'NNPS'}:
    #         result.append(word)

    result = [word for word, tag in tagged_text if tag in {'NN', 'NNS', 'NNP', 'NNPS'} ]
    
    return result
```

```python
nltk_noun('Hello My name is HyunJin')
```

##### ♦️2. 실습예제
  - 영어 불용어 적용 명사 define 만들기

```python
def extract_noun_english(string):
    string = string.lower()
    split_text = nltk.word_tokenize(string)
    tagged_text = nltk.pos_tag(split_text)

    # result = [word for word, tag in tagged_text if word not in stopwords if tag in {'NN', 'NNS', 'NNP', 'NNPS'}]

    result = []
    for word, tag in tagged_text:
        if word not in stopwords:
            if tag in {'NN', 'NNS', 'NNP', 'NNPS'}:
                result.append(word)
    
    return result
```

```python
extract_noun_english('I am Happy girl')
```

##### ♦️3. 실습예제
  - 한국어 불용어 적용된 명사/형용사/동사를 추출하는 define 만들기

```python
#원하는 태그 추출 & 불용어 제거
stopwords = ['있다' ,'되다', '하다', '나']

def okt_pos_tagging(string):
    pos_words = okt.pos(string, stem=True, norm=True)
    words = [word for word, tag in pos_words if tag in {'Noun', 'Adjective', 'Verb'} if word not in stopwords ]
    return words
```
