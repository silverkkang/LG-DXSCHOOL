```python
import pandas as pd
import pickle
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
with open('clustering_result_project.pkl', 'rb') as f:
    df = pickle.load(f)
df
```

```python
import gensim
from gensim import corpora, models
from gensim.corpora import Dictionary
```

```python
df_cluster0 = df[df['cluster']==0]
df_cluster0.head()
```

```python
all_documents = list(df_cluster0['review_words'])
dictionary = Dictionary(all_documents)
dictionary.token2id
```

```python
corpus = []

for doc in all_documents:
    corpus.append(dictionary.doc2bow(doc))

corpus[0]
```

```python
#3개의 토픽을 갖도록 LDA Modeling

topic_num =  3

ldamodel =  gensim.models.ldamodel.LdaModel(corpus,
                                            num_topics = topic_num,
                                            id2word=dictionary,
                                            passes=20,
                                            iterations=50,
                                            random_state=42
                                            )
```

```python
ldamodel.print_topics(num_words=5)
```

```python
#다큐먼트별로 토픽의 분포 확인
for i in range(0,100):
    print(ldamodel.get_document_topics(corpus[i]))
```

```python
from gensim.models import CoherenceModel #coherence 모델 라이브러리
import matplotlib.pyplot as plt #그래프 그리기
import numpy as np
```

```python
#perplexity
perplexity_score = ldamodel.log_perplexity(corpus)
perplexity_score
```

```python
#coherence
top_n = 3

coherence_model = CoherenceModel(model=ldamodel,
                                 texts = all_documents,
                                 dictionary=dictionary,
                                 topn=top_n
                                 )

coherence_score = coherence_model.get_coherence()
coherence_score
```

```python
perplexity_values = []

for i in range(2,10):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=i, id2word=dictionary)
    perplexity_v = ldamodel.log_perplexity(corpus)
    perplexity_values.append(perplexity_v)
```

```python
x = range(2,10)

plt.plot(x, perplexity_values)
plt.xlabel('Number of Topics')
plt.ylabel('Perplexity score')
```

```python
coherence_values=[]
top_n = 3

for i in tqdm(range(2,10)):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=i, id2word=dictionary)
    coherence_model = CoherenceModel(model=ldamodel,
                                 texts = all_documents,
                                 dictionary=dictionary,
                                 topn=top_n
                                 )
    coherence_v = coherence_model.get_coherence()
    coherence_values.append(coherence_v)
```

```python
x = range(2,10)

plt.plot(x, coherence_values)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Score')
```

```python
topic_num =  3

ldamodel =  gensim.models.ldamodel.LdaModel(corpus,
                                            num_topics = topic_num,
                                            id2word=dictionary,
                                            passes=20,
                                            iterations=50,
                                            random_state=42
                                            )
```

```python
ldamodel.get_document_topics(corpus)[0]
```

```python
action_align = []

for doc in tqdm(ldamodel.get_document_topics(corpus)):
    label =[]
    value =[]
    for score in doc:
        label.append(score[0])
        value.append(score[1])
    max_index = np.argmax(value)
    action_n = label[max_index]
    action_align.append(action_n)
```

```python
action_align[:20]
```

```python
df_cluster0['action_cluster'] = action_align
df_cluster0
```

```python
df_cluster0['action_cluster'].value_counts()
```
