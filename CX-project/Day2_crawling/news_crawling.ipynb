```python
#1)url, request
url1 = 'https://news.naver.com/breakingnews/section/103/241'
res= req.get(url1)



#2)Lcml
html=bs(res.text, 'lxml')
```

```python
news_list = html. select('a.sa_text_title')
news_list
```

```python
news_list[0].text
```

```python
for i in range(len(news_list)):
    print(news_list[i].text)
```

```python
news_list[0].get('href')
```

```python
href_list =[]
for i in range(len(news_list)):
    href_list.append(news_list[i].get('href'))

href_list
```

```python
for href in href_list:
    
    #url 요청
    url = href 
    res = req.get(url)
    
    #lxml파싱
    html=bs(res.text, 'lxml')
    
    #title/content
    title = html.select_one('#title_area>span')
    content = html.select_one('#dic_area').text
    
    #print
    print('뉴스제목:', title)
    print('내용:', content)
    print("="*300)
```

```python
data_list = []
for href in href_list:    
    #url 요청
    url = href 
    res = req.get(url)
    
    #lxml파싱
    html=bs(res.text, 'lxml')
    
    #title/content
    title = html.select_one('#title_area>span')
    content = html.select_one('#dic_area').text.strip()
    
    #print
    data_list.append([title, content, href])
```

```python
data_list
```

```python
import pandas as pd
```

```python
news_df = pd.DataFrame(data_list, columns=['뉴스제목','내용','페이지주소'])
news_df.head()
```

```python
news_df['내용'] = news_df['내용'].str.replace('\n','')
news_df
```
