```python
# ìˆ˜ì§‘í•œ ë°ì´í„°ë¥¼ ë‹´ì„ ë¦¬ìŠ¤íŠ¸
df = []
# ì‹¤íŒ¨í•œ URLê³¼ ì˜¤ë¥˜ ì‚¬ìœ ë¥¼ ì €ì¥í•  ë¹ˆ ë¦¬ìŠ¤íŠ¸
failed_url = []  

# ë¸Œë¼ìš°ì € ì‹¤í–‰í•˜ê¸°
driver = wb.Chrome()

# href_list ìˆœíšŒí•˜ë©° ë°ì´í„° ìˆ˜ì§‘í•˜ê¸°
for href in href_list:
    try:
        # 1) í˜ì´ì§€ ìš”ì²­í•˜ê¸°
        driver.get(href)
        # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°

        # 2) iframeìœ¼ë¡œ ì „í™˜í•˜ê¸°
        driver.switch_to.frame('mainFrame')

        # 3) ì œëª© ì¶”ì¶œí•˜ê¸°
        title = driver.find_element(By.CLASS_NAME, 'se-fs-.se-ff-nanummaruburi').text
	
        # 4) ë‚´ìš© ì¶”ì¶œí•˜ê¸°
        content = driver.find_element(By.CLASS_NAME, 'se-main-container').text

        # 5) ë‚ ì§œ ì¶”ì¶œí•˜ê¸°
        date = driver.find_element(By.CSS_SELECTOR,'.se_publishDate.pcol2').text

        # 6) df ë¦¬ìŠ¤íŠ¸ì— ì œëª©, ë‚´ìš©, ë‚ ì§œ, URL ì¶”ê°€í•˜ê¸°
        df.append([title, content, date, href])

    except Exception as e:
        # ì˜¤ë¥˜ ë°œìƒ ì‹œ failed_urlì— hrefì™€ ì—ëŸ¬ ì‚¬ìœ  ì¶”ê°€í•˜ê¸°
        failed_url.append([href, str(e)])

# 7) ë“œë¼ì´ë²„ ì¢…ë£Œí•˜ê¸°
driver.quit()

# ê²°ê³¼ í™•ì¸
df, failed_url
```

```python
import pandas as pd

blog_df = pd.DataFrame(df, columns=['ì œëª©','ë‚´ìš©','ë‚ ì§œ','ì£¼ì†Œ'])
blog_df.head()
```

```python
#. str.replace()
blog_df['ë‚´ìš©'] = blog_df['ë‚´ìš©'].str.replace('\n','')
blog_df
```

```python

blog_df['ë‚ ì§œ'] = blog_df['ë‚ ì§œ'].str[:-6]
blog_df
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()

def extract_nouns(text):
    #1) tag ê°’ì´ 'noun'ì´ê³  len(word)>=2 ê²ƒë“¤ë§Œ ë°˜í™˜í•˜ê¸°
    # nouns ë¦¬ìŠ¤íŠ¸ì— append(ì¡°ê±´ì„ ë§Œì¡±í•˜ë©´)
    nouns=[]
    #posë¡œ ê²°ê³¼ ê°’ì„ ë°›ì•„ë³´ë©´ : (ë‹¨ì–´, í’ˆì‚¬)
    for word, tag in okt.pos(text):
        if tag == "Noun" and len(word) >=2:
            nouns.append(word)
    return nouns
```

```python
blog_df["nouns"] = blog_df["ë‚´ìš©"].apply(extract_nouns)
blog_df
```

```python
#3. ë‹¨ì–´ë³„ ë¹ˆë„ ê³„ì‚°í•˜ê¸° (Counter)
from collections import Counter

all_nouns = []

#df["nouns"]
for nouns in blog_df["nouns"]:
    all_nouns.extend(nouns)
```

```python
freq_dict = Counter(all_nouns)
```

```python
top20 = freq_dict.most_common(20)
top20
```

```python
#ë³¼ ìš©ì–´ ì²˜ë¦¬í•˜ê¸° : 'ìŒì‹ë¬¼', 'ì²˜ë¦¬ê¸°'
stopwords= ["ì €ì†", "ë…¸í™”", "ë¡œë¶€í„°","í˜‘ì°¬","ì œê³µ"]

#ìƒˆë¡œìš´ í•„í„°ë§ë¤ ë¦¬ìŠ¤íŠ¸ ìƒì„±ì„ ìœ„í•œ ë¹ˆ ë¦¬ìŠ¤íŠ¸
filtered_nouns = []

for nouns in all_nouns:
    if nouns not in stopwords:
        filtered_nouns.append(nouns)

filtered_nouns
```
