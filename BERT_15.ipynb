### 2. Git 저장소 인증

개인(프라이빗) 저장소에 접근하려면 인증이 필요합니다. 가장 일반적인 방법은 GitHub Personal Access Token (PAT)을 사용하는 것입니다.

#### GitHub Personal Access Token (PAT) 설정:
1.  GitHub 계정에 로그인합니다.
2.  **Settings** > **Developer settings** > **Personal access tokens** > **Tokens (classic)**로 이동합니다.
3.  **Generate new token (classic)**을 클릭하고 적절한 권한(예: `repo` 범위)을 부여합니다.
4.  생성된 토큰을 복사합니다. **이 토큰은 한 번만 표시되므로 안전하게 보관해야 합니다.**

#### Colab에서 PAT 사용하기:
Colab의 **Secrets** 탭(왼쪽 패널의 열쇠 아이콘)에 PAT를 저장하는 것이 안전한 방법입니다. `GITHUB_TOKEN`과 같은 이름으로 저장하세요. 그런 다음 다음과 같이 사용할 수 있습니다.

```python
from google.colab import userdata

# Colab Secrets에서 GitHub 토큰 가져오기
# Secrets 탭에 'GITHUB_TOKEN'이라는 이름으로 토큰을 저장했다고 가정합니다.
# public repository는 토큰이 필요하지 않을 수도 있습니다.

# github_token = userdata.get('GITHUB_TOKEN')

# 토큰을 사용하여 HTTP 요청을 통해 Git 자격 증명을 설정할 수 있지만
# 일반적으로 아래의 클론 명령에서 직접 토큰을 URL에 포함하는 것이 일반적입니다.
# 예: `https://<YOUR_GITHUB_TOKEN>@github.com/username/repository.git`
```

### 4. 변경 사항 푸시 (Push) 및 풀 (Pull)

저장소를 클론한 후에는 파일 수정, 커밋, 푸시 등의 Git 작업을 수행할 수 있습니다. Colab 환경에서 `!` 접두사를 사용하여 Git 명령어를 실행할 수 있습니다.

```python
# 현재 디렉토리 상태 확인
!git status

# 변경 사항 추가 (예: 모든 변경된 파일)
# !git add .

# 변경 사항 커밋
# !git commit -m "Colab에서 첫 번째 커밋"

# 원격 저장소로 푸시
# !git push origin main # 또는 master 브랜치

# 원격 저장소의 최신 변경 사항 가져오기
# !git pull origin main # 또는 master 브랜치

print("이제 Git 명령어를 사용하여 저장소를 관리할 수 있습니다.")
```
