```python
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt

# ▼▼▼ 이 함수를 먼저 정의합니다 ▼▼▼
def find_optimal_k(corpus, dictionary, texts):
    coherence_values = []
    # 토픽 개수를 2부터 15까지 테스트해봅니다.
    k_range = range(2, 16) 
    for k in k_range:
        ldamodel = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                   id2word=dictionary,
                                                   num_topics=k,
                                                   random_state=42) # passes, iterations 등 파라미터 추가 가능
        
        coherence_model = CoherenceModel(model=ldamodel, texts=texts, dictionary=dictionary, topn=5)
        coherence_score = coherence_model.get_coherence()
        coherence_values.append(coherence_score)
        print(f"토픽 개수(K)={k}, Coherence 점수: {coherence_score:.4f}")

    # 그래프로 점수 변화 확인
    plt.plot(k_range, coherence_values)
    plt.xlabel("Number of Topics (K)")
    plt.ylabel("Coherence Score")
    plt.title("Find Optimal K")
    plt.show()
```

```python
import pandas as pd
import pickle
import gensim
from gensim import corpora
from gensim.models import CoherenceModel
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
import warnings
warnings.filterwarnings('ignore')

# 1. 데이터 로딩 및 필터링 (기존과 동일)
with open('clustering_result_project.pkl', 'rb') as f:
    df = pickle.load(f)

df_cluster1 = df[df['cluster'] == 1].copy() # .copy()를 추가하여 SettingWithCopyWarning 방지

# 2. 토픽 모델링을 위한 데이터 준비 (기존과 동일)
all_documents = list(df_cluster1['review_words'])
dictionary = corpora.Dictionary(all_documents)
corpus = [dictionary.doc2bow(doc) for doc in all_documents]

# 3. 최적의 토픽 개수(K) 찾기 (Perplexity, Coherence 동시 계산으로 효율화)
perplexity_values = []
coherence_values = []
k_range = range(2, 10)

for i in tqdm(k_range):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=i, id2word=dictionary, random_state=42)
    
    # Perplexity 계산
    perplexity_v = ldamodel.log_perplexity(corpus)
    perplexity_values.append(perplexity_v)
    
    # Coherence 계산
    coherence_model = CoherenceModel(model=ldamodel, texts=all_documents, dictionary=dictionary, topn=10)
    coherence_v = coherence_model.get_coherence()
    coherence_values.append(coherence_v)

# 4. 최적 K 찾기 위한 그래프 시각화 (두 그래프를 한번에 그려 비교하기 용이)
fig, ax1 = plt.subplots(figsize=(12, 5))

# Perplexity 그래프
ax1.set_xlabel('Number of Topics (K)')
ax1.set_ylabel('Perplexity Score', color='tab:blue')
ax1.plot(k_range, perplexity_values, color='tab:blue', marker='o', label='Perplexity')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Coherence 그래프 (Y축 공유)
ax2 = ax1.twinx()
ax2.set_ylabel('Coherence Score', color='tab:red')
ax2.plot(k_range, coherence_values, color='tab:red', marker='o', label='Coherence')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Perplexity & Coherence Scores by Number of Topics')
fig.tight_layout()
plt.show()

# 5. 그래프를 보고 최적의 K값으로 최종 모델 생성
# <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
# 중요: 위 그래프를 보고 Coherence가 가장 높고, Perplexity가 급격히 낮아지는 지점의 K값을 선택하세요.
OPTIMAL_K = 4 # 예시로 4를 선택. 실제 그래프를 보고 이 값을 직접 수정해야 합니다.
# <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

print(f"\n최종 선택된 토픽 개수(K): {OPTIMAL_K}")
final_ldamodel = gensim.models.ldamodel.LdaModel(corpus,
                                                 num_topics=OPTIMAL_K,
                                                 id2word=dictionary,
                                                 passes=20,
                                                 iterations=100, # 반복 횟수 증가로 품질 향상
                                                 random_state=42)

# 6. 각 문서에 대한 토픽 할당 (List Comprehension으로 간결화)
doc_topics = final_ldamodel.get_document_topics(corpus)
dominant_topics = [sorted(doc, key=lambda x: x[1], reverse=True)[0][0] for doc in tqdm(doc_topics)]
df_cluster1['action_cluster'] = dominant_topics

# 7. pyLDAvis 시각화 및 저장 (파일명 수정)
prepared_data = gensimvis.prepare(final_ldamodel, corpus, dictionary)
pyLDAvis.save_html(prepared_data, f'ldavis_cluster1_k{OPTIMAL_K}.html')
# pyLDAvis.display(prepared_data) # 주피터 노트북에서 바로 보려면 이 코드 사용

# 8. 최종 결과물 정리 및 저장
# 토픽 ID를 0,1,2... 에서 1,2,3... 으로 변경 (for문 대신 .map() 사용)
topic_mapping = {i: i + 1 for i in range(OPTIMAL_K)}
df_cluster1['action_cluster'] = df_cluster1['action_cluster'].map(topic_mapping)

final_df = df_cluster1[['review', 're_review', 'review_words', 'action_cluster']]
final_df.to_csv(f'cluster_project_1_k{OPTIMAL_K}.csv', encoding='utf-8-sig', index=False)

with open(f'Cluster1_k{OPTIMAL_K}.pkl', 'wb') as f:
    pickle.dump(final_df, f)

print("\n분석 완료! 결과 파일이 저장되었습니다.")
print(final_df['action_cluster'].value_counts().sort_index())
```

```python
# create_model.py
import pickle
from konlpy.tag import Okt
from gensim import corpora
from gensim.models import LdaModel

print("테스트용 LDA 모델과 사전을 생성합니다...")

# 예제 뉴스 기사 데이터 (3개의 토픽을 가짐)
sample_docs = [
    "정부가 집값 안정을 위해 부동산 규제 완화 등 추가 공급 대책을 발표했습니다. 시장 반응은 엇갈립니다.",
    "한국 축구 대표팀이 월드컵 최종예선에서 승리하며 본선 진출에 대한 기대감을 높였습니다. 손흥민 선수의 활약이 돋보였습니다.",
    "최근 주식 시장은 글로벌 경제 불확실성으로 인해 변동성이 커지고 있습니다. 특히 금리 인상 여부가 주요 변수입니다."
]

# 1. 텍스트 전처리 (명사 추출)
okt = Okt()
tokenized_docs = [okt.nouns(doc) for doc in sample_docs]

# 2. Gensim 사전 및 코퍼스 생성
dictionary = corpora.Dictionary(tokenized_docs)
corpus = [dictionary.doc2bow(text) for text in tokenized_docs]

# 3. LDA 모델 학습
NUM_TOPICS = 3
ldamodel = LdaModel(corpus, num_topics=NUM_TOPICS, id2word=dictionary, passes=15, random_state=42)

# 4. 모델 및 사전 파일로 저장
ldamodel.save('ldamodel.model')
with open('dictionary.pkl', 'wb') as f:
    pickle.dump(dictionary, f)

print("모델(ldamodel.model)과 사전(dictionary.pkl) 생성이 완료되었습니다.")
```
